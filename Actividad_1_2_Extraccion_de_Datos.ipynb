```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturacion Bueno.xlsx')
```

```python
df.info()
```

```python
df.head()
```

```python
#Filtro1: valores de 1000 a 2000
filtro1=df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"]<=2000)]
filtro1
```

```python
#Filtro 2
filtro2=df[~(df["CVE_VEND"] == 5.0) & ~(df["CVE_VEND"] == 4.0)]
filtro2
```

```python
#Filtro 3
```

```python
#Filtro 4
filtro4= df[(df['CAN_TOT'] < 5951.7) | (df['STATUS']== "E")]
filtro4
```

```python
#Filtro 5
filtro5=df.loc[ : , ["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
filtro5
```

```python
filtro6= df.iloc[7001:7100,: ]
filtro6
```
